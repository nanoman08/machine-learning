```python
import matplotlib.pyplot as plt
import itertools
import warnings

%matplotlib inline
```

```python
fontsizes = itertools.cycle([8, 16, 24, 32])

def example_plot(ax):
    ax.plot([1, 2])
    ax.set_xlabel('x-label', fontsize=next(fontsizes))
    ax.set_ylabel('y-label', fontsize=next(fontsizes))
    ax.set_title('Title', fontsize=next(fontsizes))

fig, ax = plt.subplots()
example_plot(ax)
plt.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
example_plot(ax1)
example_plot(ax2)
plt.tight_layout()

fig, axes = plt.subplots(nrows=3, ncols=3)
for row in axes:
    for ax in row:
        example_plot(ax)
plt.tight_layout()


fig = plt.figure()
```

```python
ax1 = plt.subplot(221)
ax2 = plt.subplot(223)
ax3 = plt.subplot(122)

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)

plt.tight_layout()
```

```python
fig = plt.figure()

ax1 = plt.subplot2grid((4, 4), (0, 0), colspan=3, rowspan=2)
ax2 = plt.subplot2grid((4, 4), (0, 3), rowspan=2)
ax3 = plt.subplot2grid((4, 4), (2, 0), colspan=1, rowspan=2)
ax4 = plt.subplot2grid((4, 4), (2, 1), rowspan=2, colspan=2)

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)

plt.tight_layout()
```
